pie chart

```python
list=[10,20,45,34,30,29]
list_item=['choco','chips','biscuit','numkin','cold drinks','cakes']
```

```python
plt.pie(list,labels=list_item)
plt.show()
```

```python
plt.pie(list,labels=list_item,autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(list,labels=list_item,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.pie(list,labels=list_item,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

```python
plt.pie(list,labels=list_item,autopct='%1.1f%%',shadow=True,radius=1.5)
plt.axis('equal')
plt.show()
```

```python
plt.pie(list,labels=list_item,autopct='%1.1f%%',shadow=True,radius=1.5,explode=[0,0.2,0,0,0,0])
plt.axis('equal')
plt.show()
```

Bar Plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='y')

plt.title('Bar Graph')
plt.ylabel('y axis')
plt.xlabel('x axis')

plt.show()
```

```python
company=['google','amazon','microsoft','fb']
revenue=[90,136,89,27]
profit=[40,5,34,12]
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='profit')
plt.title('US TECHNOLOGY STOCKS')
plt.legend()
plt.show()
```

```python
company=['google','amazon','microsoft','fb']
revenue=[90,136,89,27]
profit=[40,5,34,12]
xpos=np.arange(len(company))
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')

plt.xticks(xpos,company)
plt.title('US TECHNOLOGY STOCKS')
plt.legend()
plt.show()
```

```python
ecommerce=['myntra','snapdeal','Alibaba','Amazon','Flipkart']
Q1_profit=[35,45,100,70,40]
Q2_profit=[38,40,105,65,45]
Q3_profit=[30,42,120,72,50]
Q4_profit=[25,34,115,60,48]

plt.figure(1,figsize=(10,10))

plt.subplot(221)
plt.bar(ecommerce,Q1_profit)
plt.title('Quarter1 Profit')

plt.subplot(222)
plt.bar(ecommerce,Q2_profit)
plt.title('Quarter2 Profit')

plt.subplot(223)
plt.bar(ecommerce,Q3_profit)
plt.title('Quarter3 Profit')

plt.subplot(224)
plt.bar(ecommerce,Q4_profit)
plt.title('Quarter4 Profit')

plt.suptitle('Profit on Quarter Basic')
plt.show()
```
